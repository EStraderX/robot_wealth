```r
library('tidyquant', quietly = TRUE)
DATA_FOLDER <- '~/Resilio Sync/FXBootcamp/'
source(paste0(DATA_FOLDER,'/Code/tools/r-tools/data-utils.R'))

ASSET_LIST <- "AssetsDWX-FX-USD"
asset_list <- get_asset_list(ASSET_LIST)
```

```r
asset_list
```

```r
prices_df <- get_daily_OHLC(asset_list$Name)
head(prices_df)
```

```r
usd_prices_df <- convert_common_quote_currency(prices_df, quote_currency = 'USD')
```

```r
currencies <- get_unique_currencies(usd_prices_df)
policy_rates_df <- get_policy_rates(currencies)
tail(policy_rates_df)
```

```r
usd_extended_prices_df <- append_interest_rate_differential(usd_prices_df, policy_rates_df)
```

```r
tickers <- distinct(usd_extended_prices_df, Ticker)[[1]]
for (i in tickers) {
  print(usd_extended_prices_df %>%
      filter(Ticker == i) %>%
      select(Ticker, Date, Spot_Return_Index, Interest_Return_Index, Total_Return_Index) %>%
      gather('Spot_Return_Index','Interest_Return_Index','Total_Return_Index', key = 'type', value = 'value') %>%
      ggplot(aes(x = Date, y = value, colour = type)) +
      geom_line() +
      labs(title = i) + 
      theme_tq())
}
```

### What differentiates the EM currencies (MXN, TRY) from the developed countries?

* Emerging market interest rates like MXN and TRY are much higher than those in developed markets. 
* We can see that despite the higher interest rates in the emerging markets, it is not a strong factor and unable to drive higher/stronger fx rates for the currency. 
* This is totally counterintuitive to 'standard' economics.

### How does the volatility of exchange rates and interest rates compare?

* Increased interest rate volatility increases  exchange rate volatility. 
* Interest rate shifts, do affect some increase/decrease in exchange rate.
* We can validate this empirically by looking at changes in fx volatility before and following increases in exchange rate shifts.
